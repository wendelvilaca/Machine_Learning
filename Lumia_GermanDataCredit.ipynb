# 1 - Resumo da análise de risco de crédito da empresa Lumia

    Este resumo aglutinou as análises mais importantes desse projeto com objetivo de responder as necessidades da empresa Lunia.
    
    O algoritimo implementado teve sucesso de 73% para previsões de empréstimos bons. Nos dados analisados, a idade média das pessoas é 35,54. O valor médio do crédito emprestado é 3.271. A duração dos empréstimos em média são 20 meses. A média de anos em que os clientes moram na mesma residência é de 2,8 anos. 
    
    Há 2x mais homens do que mulheres em nosso conjunto de dados. A maioria das mulheres possuem menos de 30 anos. A maioria dos homens possuem entre 20 e 40 anos. As mulheres são mais propensas a solicitar um empréstimo de crédito para comprar móveis e equipamentos. (10% a mais que homens)
    
    Os homens utilizam o crédito duas vezes mais do que as mulheres para investir em um negócio. O dobro de mulheres estavam desempregadas em comparação com os homens. O grupo de jovens tende a pedir empréstimos ligeiramente mais elevados em comparação com os grupos de idade mais avançada.
    
    Os grupos de jovens e idosos apresentaram a maior proporção de empréstimos de alto risco. Com 45,29% de todos os clientes pertencentes à faixa etária jovem sendo considerados de alto risco. O número de empréstimos considerados de alto risco no grupo de idosos é de 44,28% do total de pessoas consideradas no grupo de idosos. No lado do grupo de idosos, este é o grupo que mais provavelmente está recebendo o dinheiro de suas pensões, o que significa que o grupo de idosos está provavelmente desempregado ou trabalhando em meio período. Idosos que solicitaram empréstimos acima de 12 mil têm uma grande chance de se tornarem empréstimos inadimplentes

    Indivíduos pertencentes a classe econômica "D" apresentaram maior probabilidade de serem créditos de risco duvidoso do que outros tipos de grupos. Quanto maior a riqueza, menor a probabilidade de ser um empréstimo de risco ruim. Quanto maior o valor do crédito, maior a probabilidade do empréstimo não ser cumprido. Quanto maior a duração do empréstimo, maior a probabilidade de não ser cumprido.
          
    Se o valor do crédito emprestado for equivalente a 11.000 ou mais, a probabilidade de o empréstimo ser ruim aumenta drasticamente. Automóveis, Rádio / TV e Móveis e Equipamentos respondem por mais de 50% do risco total e tem a maior distribuição de crédito concedido. Os objetivos  de crédido de carros e rádio / TV eram os menos arriscados do ponto de vista operacional, uma vez que apresentava a maior lacuna entre o risco bom e o mau. 

    Conforme supracitado, o algoritimo obteve sucesso de 73% nas previsões de crédito. Para melhoria seria necessário manipular mais variáveis e identificar outros apdrões bem como testar outros algoritmos considerando os grupos. Como melhoria de captção de dados se fosse possível obter mais informações dos clientes referente aos bens de consumo do mesmo provavelmente seria mais assertivo as previsões.

```python
#3.2 Seleção e tradução dos dados


StatusDaContaCorrenteExistente={'A14':"Sem conta",'A11':"Saldo negativo", 'A12': "Saldo positivo até R$200",
                                'A13':"Saldo maior que R$200"}
df_credit1["StatusDaContaCorrenteExistente"]=df_credit1["StatusDaContaCorrenteExistente"].map(StatusDaContaCorrenteExistente)

Credit_history={"A34":"conta crítica(Dívidas em outros bancos)","A33":"atraso de pagamento no passado",
                "A32":"créditos existentes reembolsados devidamente até agora",
                "A31":"todos os créditos neste banco reembolsados devidamente",
                "A30":"nenhum crédito recebido/reembolsados devidamente"}
df_credit1["HistoricoDeCredito"]=df_credit1["HistoricoDeCredito"].map(Credit_history)

Objetivo={"A40" : "carro (novo)", "A41" : "carro (usado)", "A42" : "móveis / equipamentos", "A43" :"rádio / televisão" , 
          "A44" : "eletrodomésticos", "A45" : "reparos", "A46" : "educação", 'A47' : 'férias','A48' : 'retreinamento',
          'A49' : 'negócios','A410' : 'outros'}
df_credit1["Objetivo"]=df_credit1["Objetivo"].map(Objetivo)

Saving_account={"A65" : "sem conta poupança","A61" :"D","A62" : "C",
                "A63" :"B", "A64" :"A"}
df_credit1["PoupancaTitulos"]=df_credit1["PoupancaTitulos"].map(Saving_account)

Present_employment={'A75':">=7 anos", 'A74':"4<= <7 anos",  'A73':"1<= < 4 anos", 'A72':"<1 anos",'A71':"desempregado"}
df_credit1["EmpregoAtualDesde"]=df_credit1["EmpregoAtualDesde"].map(Present_employment)


#Descrição original, antes de alterar apra somente masculino e feminino
#Personal_status_and_sex={ 'A95':"female:single",'A94':"male:married/widowed",
#                         'A93':"male:single", 'A92':"female:divorced/separated/married", 'A91':"male:divorced/separated"}
#df_credit1["Personal status and sex"]=df_credit1["Personal status and sex"].map(Personal_status_and_sex)

Personal_status_and_sex={ 'A95':"feminino",'A94':"masculino",
                         'A93':"masculino", 'A92':"feminino", 'A91':"masculino"}
df_credit1["StatusPessoalESexo"]=df_credit1["StatusPessoalESexo"].map(Personal_status_and_sex)

Other_debtors_guarantors={'A101':"nenhum", 'A102':"co-requerente", 'A103':"fiador"}
df_credit1["OutrosDevedoresFiadores"]=df_credit1["OutrosDevedoresFiadores"].map(Other_debtors_guarantors)

Property={'A121':"imóveis", 'A122':"contrato de poupança / seguro de vida", 'A123':"carro ou outro",
          'A124':"desconhecido"}
df_credit1["Propriedade"]=df_credit1["Propriedade"].map(Property)

Other_installment_plans={'A143':"nenhum", 'A142':"lojas", 'A141':"banco"}
df_credit1["OutrosPlanosDeParcelamento"]=df_credit1["OutrosPlanosDeParcelamento"].map(Other_installment_plans)

Housing={'A153':"gratuita", 'A152':"própria", 'A151':"aluguel"}
df_credit1["Habitacao"]=df_credit1["Habitacao"].map(Housing)

Job={'A174':"gestor / altamente qualificado", 'A173':"funcionário / oficial qualificado", 
     'A172':"não qualificado - residente", 'A171':"desempregado"}
df_credit1["Emprego"]=df_credit1["Emprego"].map(Job)

Telephone={'A192':"yes", 'A191':"none"}
df_credit1["Telefone"]=df_credit1["Telefone"].map(Telephone)

foreign_worker={'A201':"yes", 'A202':"no"}
df_credit1["TrabalhadorEstrangeiro"]=df_credit1["TrabalhadorEstrangeiro"].map(foreign_worker)

risk={1:"Bom", 2:"Ruim"}
df_credit1["RiscoDeCredito"]=df_credit1["RiscoDeCredito"].map(risk)
```
